```python
import numpy as np
from neurons import Neuron, simulate
import matplotlib.pyplot as plt
```

```python
T = 200  # simulation duration in milliseconds
dt = 0.01 # time step in millisecond
time_axis = np.arange(0, T, dt)
```

```python
v_init = -65
u_init = 1.3
I = 70
a = 0.02
b = 0.2
c = -55
d = 8
neuron = Neuron(v_init, u_init, I, a, b, c, d)
```

```python
record = simulate(neuron, dt, T/dt)
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,4))
plt.plot(time_axis, record)
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Membrane Potential (mV)')
plt.show()
```
