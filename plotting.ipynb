```python
import numpy as np
from neurons import Neuron, simulate
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
T = 500  # simulation duration in milliseconds
dt = 0.1 # time step in millisecond
N = int(T / dt)
time_axis = np.arange(0, T, dt)
```

```python
v_init = -70
u_init = 0
I = np.full(N, 20.0)
I[:1000] = 0.0
I[-1000:] = 0.0
a = 0.02
b = 0.2
c = -55
d = 4
neuron = Neuron(v_init, u_init, a, b, c, d)
```

```python
record = simulate(neuron, I, dt)
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,4))
plt.plot(time_axis, 0.1*I-100)
plt.plot(time_axis, record)
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Membrane Potential (mV)')
plt.show()
```
